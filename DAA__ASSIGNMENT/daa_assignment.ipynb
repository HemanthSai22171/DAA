```python
# 1

# Iterative Algorithm

import time

def iterative_sum(n):
    result = 0
    for i in range(1, n+1):
        result += i
    return result

start_time = time.time()
result_iterative = iterative_sum(1000) 
end_time = time.time()
execution_time_iterative = end_time - start_time
print("Iterative Sum:", result_iterative)
print("Execution Time (Iterative):", execution_time_iterative, "seconds")

# Recursive Algorithm

def recursive_sum(n):
    if n == 0:
        return 0
    else:
        return n + recursive_sum(n-1)

start_time = time.time()
result_recursive = recursive_sum(1000)  
end_time = time.time()
execution_time_recursive = end_time - start_time
print("Recursive Sum:", result_recursive)
print("Execution Time (Recursive):", execution_time_recursive, "seconds")
```

```python
import random
import time
import matplotlib.pyplot as plt

arr = [random.randint(1, 1000) for _ in range(10000)]

# linear search function
def linear_search(arr, key):
    start_time = time.time()
    for i in range(len(arr)):
        if arr[i] == key:
            end_time = time.time()
            return end_time - start_time
    return -1

# binary search function
def binary_search(arr, key):
    start_time = time.time()
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == key:
            end_time = time.time()
            return end_time - start_time
        elif arr[mid] < key:
            low = mid + 1
        else:
            high = mid - 1
    return -1

search_keys = [random.randint(1, 1000) for _ in range(5)]
execution_times_linear = []
execution_times_binary = []

for key in search_keys:
    execution_times_linear.append(linear_search(arr, key))
    execution_times_binary.append(binary_search(arr, key))

plt.plot(search_keys, execution_times_linear, label='Linear Search')
plt.plot(search_keys, execution_times_binary, label='Binary Search')
plt.xlabel('Search Key')
plt.ylabel('Execution Time (seconds)')
plt.title('Execution Time of Linear and Binary Searches')
plt.legend()
plt.show()
```

```python
# 3

def string_to_int(s):
    if len(s) == 1:
        return int(s)
    else:
        return int(s[0]) * 10**(len(s)-1) + string_to_int(s[1:])

result_int = string_to_int("13531")
print("Converted Integer:", result_int)
```

```python
# 4

def reverse_string(s):
    if len(s) == 0:
        return s
    else:
        return reverse_string(s[1:]) + s[0]

result_reverse = reverse_string("pots&pans")
print("Reversed String:", result_reverse)
```

```python
# 5

def is_palindrome(s):
    if len(s) <= 1:
        return True
    elif s[0] != s[-1]:
        return False
    else:
        return is_palindrome(s[1:-1])

print(is_palindrome("racecar")) 
print(is_palindrome("gohangasalamiimalasagnahog"))  
print(is_palindrome("hello")) 
```

```python
# 1

import random
import time
import matplotlib.pyplot as plt

random_numbers = [random.randint(1, 10000) for _ in range(1000)]

# Bubble Sort
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]

# Selection Sort
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]

# Insertion Sort
def insertion_sort(arr):
    n = len(arr)
    for i in range(1, n):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

# Merge Sort
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        L = arr[:mid]
        R = arr[mid:]

        merge_sort(L)
        merge_sort(R)

        i = j = k = 0

        while i < len(L) and j < len(R):
            if L[i] < R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1

        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1

        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1

# Quick Sort
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    else:
        pivot = arr[0]
        lesser = [x for x in arr[1:] if x <= pivot]
        greater = [x for x in arr[1:] if x > pivot]
        return quick_sort(lesser) + [pivot] + quick_sort(greater)

sorting_algorithms = {
    'Bubble Sort': bubble_sort,
    'Selection Sort': selection_sort,
    'Insertion Sort': insertion_sort,
    'Merge Sort': merge_sort,
    'Quick Sort': quick_sort
}

execution_times = {}

for algo_name, algo_func in sorting_algorithms.items():
    arr_copy = random_numbers.copy()
    start_time = time.time()
    algo_func(arr_copy)
    end_time = time.time()
    execution_times[algo_name] = end_time - start_time

plt.bar(range(len(execution_times)), list(execution_times.values()), align='center')
plt.xticks(range(len(execution_times)), list(execution_times.keys()), rotation=45)
plt.xlabel('Sorting Algorithm')
plt.ylabel('Execution Time (seconds)')
plt.title('Comparison of Sorting Algorithms')
plt.show()
```

```python
# 2

import heapq

def print_sorted_lists(lists):
    result = []
    heap = [(lst[0], i, 0) for i, lst in enumerate(lists) if lst]
    heapq.heapify(heap)

    while heap:
        val, list_idx, idx = heapq.heappop(heap)
        result.append(val)
        if idx + 1 < len(lists[list_idx]):
            next_val = lists[list_idx][idx + 1]
            heapq.heappush(heap, (next_val, list_idx, idx + 1))

    print(result)

# Example usage
sorted_lists = [[10, 20, 30, 40], [15, 25, 35], [27, 29, 37, 48, 93], [32, 33]]
print_sorted_lists(sorted_lists)
```

```python
# 3

import heapq

def find_k_largest(arr, k):
    min_heap = []
    for num in arr:
        if len(min_heap) < k:
            heapq.heappush(min_heap, num)
        else:
            if num > min_heap[0]:
                heapq.heappop(min_heap)
                heapq.heappush(min_heap, num)
    return min_heap

arr = [10, 15, 20, 25, 27, 29, 30, 32, 33, 35, 37, 40, 48, 93]
k = 3
result = find_k_largest(arr, k)
print(result)
```

```python
# 4

def max_activities(activities):
    activities.sort(key=lambda x: x[1]) 
    selected = [activities[0]]
    last_finish = activities[0][1]

    for activity in activities[1:]:
        start, finish = activity
        if start >= last_finish:
            selected.append(activity)
            last_finish = finish

    return selected

activities = [(1, 4), (3, 5), (0, 6), (5, 7), (3, 8), (5, 9), (6, 10), (8, 11), (8, 12), (2, 13), (12, 14)]
result = max_activities(activities)
print(result)
```

```python
# 5

def merge_intervals(intervals):
    intervals.sort(key=lambda x: x[0])  # Sort by start time
    merged = [intervals[0]]

    for interval in intervals[1:]:
        if interval[0] <= merged[-1][1]:
            merged[-1] = (merged[-1][0], max(merged[-1][1], interval[1]))
        else:
            merged.append(interval)

    return merged

# Example usage
intervals = [(1, 4), (2, 5), (7, 8), (6, 9)]
result = merge_intervals(intervals)
print(result)
```

```python
# 1

def find_pair_with_sum(arr, target_sum):
    seen = set()
    for num in arr:
        complement = target_sum - num
        if complement in seen:
            return num, complement
        seen.add(num)
    return None

arr = [8, 7, 2, 5, 3, 1]
target_sum = 10
result = find_pair_with_sum(arr, target_sum)
print(result)
```

```python
# 2

def find_max_product_pair(arr):
    max_product = float('-inf')
    min_product = float('inf')
    max_pair = None

    for num in arr:
        if num >= 0:
            max_product = max(max_product, num * max_product if max_product != float('-inf') else num)
            min_product = min(min_product, num * min_product if min_product != float('inf') else num)
        else:
            temp = max_product
            max_product = max(min_product * num, num)
            min_product = min(temp * num, num)

        if max_product > min_product and max_product > num * min_product:
            max_pair = (max_product, num)

    return max_pair

arr = [1, 7, 4, 2, 8, 6, 3, 9, 5]
result = find_max_product_pair(arr)
print(result)
```

```python
# 3

def sort_array_with_swapped_elements(arr):
    n = len(arr)
    first_swap = None
    second_swap = None

    for i in range(n - 1):
        if arr[i] > arr[i + 1]:
            if first_swap is None:
                first_swap = i
            else:
                second_swap = i + 1

    if second_swap is None:
        second_swap = first_swap + 1

    arr[first_swap], arr[second_swap] = arr[second_swap], arr[first_swap]

    return arr

arr = [3, 8, 6, 7, 5, 9]
result = sort_array_with_swapped_elements(arr)
print(result)
```

```python
# 4

def segregate_zeros_and_ones(arr):
    left = 0
    right = len(arr) - 1

    while left < right:
        while arr[left] == 0 and left < right:
            left += 1
        while arr[right] == 1 and left < right:
            right -= 1
        if left < right:
            arr[left], arr[right] = arr[right], arr[left]
            left += 1
            right -= 1

    return arr

arr = [0, 1, 0, 1, 0, 0, 1, 1, 1, 0]
result = segregate_zeros_and_ones(arr)
print(result)
```

```python
# 5

def count_inversions(arr):
    if len(arr) <= 1:
        return arr, 0

    mid = len(arr) // 2
    left, left_inv = count_inversions(arr[:mid])
    right, right_inv = count_inversions(arr[mid:])
    merged, split_inv = merge(left, right)

    return merged, left_inv + right_inv + split_inv

def merge(left, right):
    result = []
    split_inv = i = j = 0

    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
            split_inv += len(left) - i

    result.extend(left[i:])
    result.extend(right[j:])

    return result, split_inv

arr = [10, 1, 2, 4, 13, 9, 5]
sorted_arr, inv_count = count_inversions(arr)
print("Sorted Array:", sorted_arr)
print("Inversion Count:", inv_count) 
```

```python
# 6

# a. To solve this problem in O(n^2) time, we can use nested loops to check every pair of numbers in the array :

def has_sum_n2(arr, target):
    for i in range(len(arr)):
        for j in range(i + 1, len(arr)):
            if arr[i] + arr[j] == target:
                return True
    return False

arr = [8, 4, 1, 6]
target = 10
result = has_sum_n2(arr, target)
print(result)
```

```python
# 6

# b. To solve this problem in O(n log n) time, we can sort the array first and then use two pointers to find the pair :

def has_sum_nlogn(arr, target):
    arr.sort()
    left = 0
    right = len(arr) - 1

    while left < right:
        current_sum = arr[left] + arr[right]
        if current_sum == target:
            return True
        elif current_sum < target:
            left += 1
        else:
            right -= 1

    return False

arr = [8, 4, 1, 6]
target = 10
result = has_sum_nlogn(arr, target)
print(result)
```
